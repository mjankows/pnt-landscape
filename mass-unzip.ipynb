```python
import gzip
import os
import json
import numpy as np
import pandas as pd
from collections import defaultdict, Counter
```

```python
res = []
for it in os.listdir('openalex-snapshot'):
    for it2 in os.listdir(f'openalex-snapshot/{it}/'):
        with gzip.open(f'openalex-snapshot/{it}/{it2}','rb') as f:
            for line in f:
                try:
                    res.append(line.decode('utf-8').strip())
                except:
                    pass
```

```python
resDict = [json.loads(it) for it in res]
```

```python
resConcepts = []
for it in os.listdir('openalex-snapshot-concepts'):
    for it2 in os.listdir(f'openalex-snapshot-concepts/{it}/'):
        with gzip.open(f'openalex-snapshot-concepts/{it}/{it2}','rb') as f:
            for line in f:
                try:
                    resConcepts.append(line.decode('utf-8').strip())
                except:
                    pass
```
